# Using this package: validate_binary_y_and_X

```validate_binary_y_and_X()``` is a function you can call to confirm that the data you are working with is compatible with the risk stratifier package.

## What is it validating?

This function expects pandas objects to be passed to it, namely ```y``` and ```X```.

It then confirms the following:

1. ```y``` is a pandas series (a vector), in integer form with 0s and 1s only. 1 is treated as the positive case.
2. ```X``` is a pandas dataframe with columns of the type numeric, integer or string.
3. Both are the same length.
4. There are more than 100 rows.

## Why is this enforced?

- Pandas objects operate welll within sklearn (the machine learning workhorse of this package).
- Requiring the same length is always necessary.
- Most algorithms work best with the binary dependent variable being in integer type (rather than boolean).
- Feature engineering should primarily be handled within the pipelines you create. Categorical and boolean type pandas columns are sometimes poorly handled by sklearn pipelines and estimators.

## What does success look like?

To use the function, first import it

````python
import pandas as pd
import numpy as np
from risk_stratifier import validate_binary_y_and_X
````

Lets generate a permissable y and X

````python
y = pd.Series(np.random.randint(0, 2, size=200), dtype="int64")
X = pd.DataFrame({
    "int_col": np.random.randint(0, 100, size=200).astype("int64"),
    "float_col": np.random.rand(200).astype("float64"),
})
````

Are these objects valid?

````python
validate_binary_y_and_X(y, X)
````

## What does failure look like?

Lets create a y that is permissable in all ways but is not of the same length

````python
y_too_long = pd.Series(np.random.randint(0, 2, size=500), dtype="int64")
validate_binary_y_and_X(y_too_long, X)
````

What about incorrectly specified X columns?

````python
X_wrong_cols = pd.DataFrame({
    "int_col": np.random.randint(0, 100, size=200).astype("int64"),
    "float_col": np.random.rand(200).astype("float64"),
    "bool_col": np.random.rand(200) > 0
})
validate_binary_y_and_X(y, X_wrong_cols)
````
